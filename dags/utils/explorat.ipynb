```python
# 1. Carregue o dado original (MUDE O CAMINHO AQUI)
path_raw = "E:\\main_folder\\Tech\\Projetos\\etl_eletronics_duckdb\\data\\bronze\\electronics_sales.csv"
df_raw = pd.read_csv(path_raw)

# 2. Carregue a dimensão de produtos (MUDE O CAMINHO SE PRECISAR)
path_dim = "E:\\main_folder\\Tech\\Projetos\\etl_eletronics_duckdb\\data\\gold\\dim_product.csv"
df_dim_product = pd.read_csv(path_dim)

print("--- INICIANDO TESTES DE VALIDAÇÃO ---")

# --- TESTE 1: CONTAGEM DE PEDIDOS ---
# MUDE 'id_da_coluna_no_original' para o nome da coluna de ID no dataset cru
pedidos_originais = df_raw['order_id'].nunique()
pedidos_gold = df['order_id'].nunique() # Esse df é o fact_order que vc já carregou

print(f"\n1. Integridade de Pedidos:")
print(f"Origem: {pedidos_originais} | Gold: {pedidos_gold}")
print("Passou?", pedidos_originais == pedidos_gold)

# --- TESTE 2: FINANCEIRO ---
# Multiplicando a coluna de quantidade pela de preço (linha a linha) e depois somando o total
receita_original = round((df_raw['quantity'] * df_raw['unit_price']).sum(), 2)
receita_gold = round(df['gross_revenue'].sum(), 2)

print(f"\n2. Integridade Financeira:")
print(f"Origem: R$ {receita_original} | Gold: R$ {receita_gold}")
print("Passou?", receita_original == receita_gold)

# --- TESTE 3: PRODUTOS ÓRFÃOS ---
produtos_na_fato = set(df['product_id'].unique())
produtos_na_dim = set(df_dim_product['product_id'].unique())
produtos_orfaos = produtos_na_fato - produtos_na_dim

print(f"\n3. Integridade Relacional (Produtos Órfãos na Fato):")
print(f"Qtd de IDs órfãos: {len(produtos_orfaos)}")
print("Passou?", len(produtos_orfaos) == 0)

```
